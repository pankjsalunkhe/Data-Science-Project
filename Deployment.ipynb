# IMPORTING  LIBRARIES


```python
import pandas as pd
import streamlit as st
from imblearn.combine import SMOTEENN
from collections import Counter
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

# CODE FOR USER INPUT FUNCTIONS IN SIDE BAR AND CREATING DATASET


```python
# Set the title for your Streamlit app
st.title('Model Development \n Telecommunication Churning')

# Create a sidebar for input features
st.sidebar.header('Input Features')

# Define a function to collect user input features
def input_features():
    # Create dropdown for Voice Plan selection (1 or 0)
    Voice_Plan=st.sidebar.selectbox('Voice Plan',('1','0'))
    # Create dropdown for International Plan selection (1 or 0)
    International_Plan=st.sidebar.selectbox('International Plan',('1','0'))
    # Create dropdown for International Plan selection (1 or 0)
    International_Calls=st.sidebar.number_input('Insert Number Of Calls')
    # Create a numeric input field for international charges
    International_Charges=st.sidebar.number_input('Insert International Charge')
    # Create a numeric input field for day charges
    Day_Charges=st.sidebar.number_input('Insert Day Charge')
    # Create a numeric input field for evening minutes
    Evening_Mins=st.sidebar.number_input('Insert Evening Minutes')
    # Create a numeric input field for night minutes
    Night_Mins=st.sidebar.number_input('Insert Night Minutes')

    # Create a dictionary to store the user input data
    data={'Voice_Plan':Voice_Plan,
          'International_Plan':International_Plan,
          'International_Calls':International_Calls,
          'International_Charges':International_Charges,
          'Day_Charges':Day_Charges,
          'Evening_Mins':Evening_Mins,
          'Night_Mins':Night_Mins}
    # Create a DataFrame with the user input data
    features=pd.DataFrame(data,index=[0])
    return features
# Call the input_features() function to collect user input
df=input_features()

# Display the user input features as a subheader and a DataFrame
st.subheader('User Input Features')
st.write(df)

```

# SPLITING DATA INTO TEST AND TRAIN AND SMOOTHENING TRAIN DATASET

```python
# Load your dataset 'Churn_Without_Outliers' using pandas
churn=pd.read_csv('Churn_Without_Outliers',encoding='utf_8')

# Extract features (X) and target (Y) variables from the dataset
x=churn.iloc[:,1:8]
y=churn.iloc[:,8]

# Split the dataset into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25)

# Apply the SMOTEENN
sm=SMOTEENN()

# Resample the training data to balance the class distribution
X,Y = sm.fit_resample(x_train,y_train)
# X contains the resampled features, and Y contains the corresponding target labels.
```

# MODEL BUILDING AND PREDICTING IF CUSTOMER WILL CHURN OR NOT

```python
# Create a RandomForestClassifier instance
RF=RandomForestClassifier()

# Train the Random Forest classifier on the resampled training data
RF.fit(X,Y)

# Make predictions on a new data point (df) using the trained model
predict=RF.predict(df)        # This will provide the predicted class (0 or 1)

# Get the prediction probabilities for each class (0 and 1)
prediction_probability=RF.predict_proba(df)

```

# PRESENTING THE RESULT ON SCREEN


```python
# Display a subheader for the prediction result
st.subheader('Prediction Result')

# Check if the predicted probability of class 1 is greater than 0.5
# If it is, display 'Yes'; otherwise, display 'No'
st.write('Yes' if prediction_probability[0][1]>0.5 else 'No')

# Display a subheader for the prediction probability
st.subheader('Prediction Probability')

# Write the prediction probability to the Streamlit app
st.write(prediction_probability)
```
